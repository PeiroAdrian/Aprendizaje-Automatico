# Regresión polinomial

## Importar Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar el dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Entrenar el modelo de regresión lineal

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

## Entrenar el modelo de regresión polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualizar los resultados de la regresión líneal

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Verdad o blofeo (regresión lineal)')
plt.xlabel('Posición salarial')
plt.ylabel('Salario')
plt.show()
```

## Visualizar los resultados de la regresión polinomial

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Verdad o blofeo (regresión polinomial)')
plt.xlabel('Posición salarial')
plt.ylabel('Salario')
plt.show()
```

## Visualización de los resultados de la regresión polinómica (para una resolución más alta y una curva más suave)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Verdad o blofeo (regresión polinomial)')
plt.xlabel('Posición salarial')
plt.ylabel('Salario')
plt.show()
```

## Predecir un nuevo resultado con regresión lineal

```python
lin_reg.predict([[6.5]])
```

## Predecir un nuevo resultado con regresión polinómica

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
